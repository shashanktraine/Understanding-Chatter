# <span style="font-size:25px;"> **Importing Libraries**

```python

import numpy as np
import pandas as pd
import math
```

# <span style="font-size:25px;"> **Folder Path**

```python

freq_file_path = r"C:\Users\shashank_p\Desktop\freq_data.xlsx"
```

# <span style="font-size:25px;"> **R Calculation**

```python




df = pd.read_excel(freq_file_path)



freq_df = df["Frequency (Hz)"]
damp_df= df["Damping percentage (%)"]
real_df = df["Real(Re)"]
imag_df = df["imaginary(Im)"]

R_list = []
for i in range(len(df)):
     freq_val = freq_df[i]
     damp_val = damp_df[i]
     real_val = real_df[i]
     imag_val = imag_df[i]

     damping_value = damp_val/100
     damping_frequency_Wd =  freq_val * np.sqrt(1-(damping_value**2))


     lambda_r_real = -1 * (damping_value * freq_val)
     lambda_r_imag = damping_frequency_Wd
     comp_lambda_r = complex(lambda_r_real,lambda_r_imag)
     con_jwr = np.conjugate(comp_lambda_r)

     comp_H_jw = complex(real_val,imag_val)


     jwr_real = 0
     jwr_imag = freq_val
     comp_jwr = complex(jwr_real,jwr_imag)

     jwd_real = 0
     jwd_imag = damping_frequency_Wd
     comp_jwd = complex(jwd_real,jwd_imag)


     jwr_lambda = comp_jwr - comp_lambda_r
     jwr_con_lambda = comp_jwr - con_jwr

     R_num = jwr_lambda * jwr_con_lambda * comp_H_jw
     R_den = (2 * comp_jwd)

     R = R_num / R_den
     R_clean = str(R).strip("()")
     R_list.append(R_clean)
R_df = pd.DataFrame({"R": R_list})

new_df = pd.concat([df,R_df],axis=1)
new_df.to_excel(r"C:\Users\shashank_p\Desktop\R_data.xlsx")
```
